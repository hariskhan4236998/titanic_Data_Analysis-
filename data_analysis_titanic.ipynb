```python

td['Survived'].value_counts().plot.pie(autopct='%1.1f%%', labels=['Not Survived', 'Survived'])
plt.title('Survival Rate')
plt.show()
```

```python
import seaborn as sns


sns.boxplot(x='Pclass', y='Fare', data=td)
plt.title('Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python

print("Average Age and Fare by Survival Status:")
print(td.groupby('Survived')[['Age', 'Fare']].mean())
```

```python

print("Passenger Distribution by Class and Gender:")
print(td.groupby(['Pclass', 'Sex'])['PassengerId'].count())
```

```python

td.fillna(td['Age'].mean(), inplace=True)


print("Missing values in 'Age' after filling:", td['Age'].isnull().sum())
```

```python

td = td.dropna(subset=['Embarked'])


print("Missing values in 'Embarked' after dropping:", td['Embarked'].isnull().sum())
```
